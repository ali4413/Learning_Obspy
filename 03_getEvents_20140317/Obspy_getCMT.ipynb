```python
import urllib
import urllib2
from collections import OrderedDict

def get_mt(yr, mo, day, oyr, omo, oday, ct_lon = None, ct_lat = None):
    locs = locals()  
    #print locs
    
    base_url = 'http://www.globalcmt.org/cgi-bin/globalcmt-cgi-bin/CMT4/form'
    
    #note here, we must use the ordered Dictionary, so that the order in the 
    #url is exactly the same order
    param = OrderedDict()
    param['itype'] = 'ymd'
    param['yr'] = yr
    param['mo'] = mo
    param['day'] = day
    param['otype'] = 'ymd'
    param['oyr'] = oyr
    param['omo'] = omo
    param['oday'] = oday
    param['jyr'] = '1976'
    param['jday'] = '1'
    param['ojyr'] = '1976'
    param['ojday'] = '1'
    
    param['nday'] = '1'
    param['lmw'] = '0'
    param['umw'] = '10'
    param['lms'] = '0'
    param['ums'] = '10'
    param['lmb'] = '0'
    param['umb'] = '10'
    
    if ct_lon and ct_lat is not None:
        llat = ct_lat - 1
        ulat = ct_lat + 1
        llon = ct_lon - 1
        ulon = ct_lon + 1
    else:
        llat = '-90'
        ulat = '90'
        llon = '-180'
        ulon = '180'
    
    param['llat'] = llat
    param['ulat'] = ulat
    param['llon'] = llon
    param['ulon'] = ulon
    
    param['lhd'] = '0'
    param['uhd'] = '1000'
    param['lts'] = '-9999'
    param['uts'] = '9999'
    param['lpe1'] = '0'
    param['upe1'] = '90'
    param['lpe2'] = '0'
    param['upe2'] = '90'
    param['list'] = '6'
    
    url = "?".join((base_url, urllib.urlencode(param)))
    print url
    return url
    
```
