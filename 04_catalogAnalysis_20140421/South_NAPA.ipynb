```python
from obspy.core.util import locations2degrees
new_dist = []
for i in range(len(lats_new)):
    s = copysign(1, lats_new - evla_1)    
    new_dist.append(s[i] *locations2degrees(lats_new[i], lons_new[i], evla_1, evlo_1)*111.2)
```
