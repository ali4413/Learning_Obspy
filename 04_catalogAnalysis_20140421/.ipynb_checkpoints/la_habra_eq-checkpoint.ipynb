```python
import warnings
warnings.filterwarnings('ignore')
%pylab inline
import pandas as pd
import cat_analysis
from obspy import UTCDateTime
#plt.xkcd
```

##Get data from USGS

```python
t1 = UTCDateTime("2014-03-28T13:59:00.000")
t2 = UTCDateTime("2014-04-02T13:00:00.000")

evla_0 = 33.919
evlo_0 = -117.944

cat = cat_analysis.get_cat(starttime = t1, endtime = t2, minmagnitude=-1, latitude=evla_0, longitude= evlo_0,
                        maxradius = 0.1,includeallorigins=True)
print cat
```

###Convert catalog data to list

```python
#Here you can choose 'Moment_tensor' or 'Focal'
earthquakes, mt, event_id, quarry_blast = cat_analysis.cat2list(cat, mt_type = 'Focal')
```

##Plot seismicity

```python
cat_analysis.plot_event(earthquakes, llat = 33.8, ulat = 34.0, llon = -118.1, 
                      ulon = -117.8, resolution='l', color = 'date', figsize = (12,8)
                      , par_range = (33.8, 34.0, 0.2), mer_range = (-117.8, -118.1, -0.2), showHour = True)
```

##Plot scatter plot

```python
df = cat_analysis.eq2df(earthquakes)
pd.scatter_matrix(df[['evla', 'evlo', 'evdp', 'mag']], figsize = (10,10), alpha=0.7, diagonal='hist')
```

##Plot seismicity rate

```python
cat_analysis.plot_seimicity_rate(earthquakes, time = 'hour')
```

##Plot cumulative number of earthquakes

```python
cat_analysis.plot_cum(earthquakes, freq = '1T')
```

##Plot MT/Focal

```python
cat_analysis.plot_mt(earthquakes, mt, event_id, radius=0.1, dist_bt=5, mt_width=0.008, show_above_M = True, M_above = 2,
                     llat = '33.7', ulat = '34.1', llon = '-118.15', ulon = '-117.75',
                     figsize = (18,16), show_eq = True, angle_step=20, legend_loc=3, pretty = False,
                     par_range = (33.8, 34.0, 0.2), mer_range = (-118.1, -117.7,  0.2))
```

##Historical MT


```python
t1 = UTCDateTime("1976-01-01")
t2 = UTCDateTime("2014-04-05")

psmeca = cat_analysis.get_hist_mt(t1, t2, evla = 33.9192, evlo= -117.944, step = 1)
cat_analysis.plot_hist_mt(psmeca, mt_size = 0.08, pretty = False)
```
