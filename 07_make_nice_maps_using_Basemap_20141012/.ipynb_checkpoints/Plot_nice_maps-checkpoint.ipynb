```python
import warnings
warnings.filterwarnings('ignore')
#%pylab inline
from obspy import UTCDateTime
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import numpy as np
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import LightSource
from mpl_toolkits.basemap import cm
```

##First let's get data from ANSS catalog

```python
def get_event_from_ANSS_file(filename, skiprows = 0):
    '''
    Funciton to read in ANSS catalog and store them into list
    Input: 
    filename - file path of your catalog
    skiprows - how many rows you want to skip
    
    return:
    list of latitude, longitude, magnitude, and times. 
    '''
    
    f=open(filename)
    myFile=f.readlines()
    f.close()
    
    lats = []
    lons = []
    mags = []
    times = []
    
    for i, line in enumerate(myFile):
        if i < skiprows:
            continue
        #time = 
        lat = float(line[24:31])
        lon = float(line[32:41])
        mag = float(line[50:54])
        dep = float(line[44:48])
        t = line[0:22]
        tt = UTCDateTime(t.replace('/', '-').replace(' ', 'T'))
        
        
        lats.append(lat)
        lons.append(lon)
        mags.append(mag)
        times.append(tt)
        
    return lats, lons, mags, times

def read_faults(input_file):
    '''
    Read in the faults file to plot the faults
    '''
    count = 0
    faults = {}
    lats = []
    lons = []
    for line in open(input_file):
        # Check for empty/commented lines
        if not line or line.startswith('>'):
            # If 1st one: new block
            if count == 0:
                pass
            count += 1
        # Non empty line: add line in current(last) block
            faults[count] = [lons, lats]
            lats = []
            lons = []
            
        else:
            lat = float(line.split()[1])
            lon = float(line.split()[0])
            lats.append(lat)
            lons.append(lon)            
            
    return faults
```

```python
lats, lons, mags, times = get_event_from_ANSS_file('./LomaPrieta_OneMonth_M2_above.txt', skiprows = 13)
```

##Separate the mainshock from the aftershocks

```python
#here we get the mainshock info, and delete them from the lists
index = np.where(mags == np.max(mags))[0][0]
main_lat = lats[index]
main_lon = lons[index]
main_mag = mags[index]
main_time = times[index]
del lats[index]
del lons[index]
del mags[index]
del times[index]
```

##Plot figure

```python
fig = plt.figure(figsize = (12,9))
map_ax = fig.add_axes([0.03, 0.13, 0.94, 0.82])
map = Basemap(projection='merc', resolution='l', lat_0 = 37, lon_0 = -121.5,
                        llcrnrlon=-122.7,llcrnrlat=36.2, urcrnrlon=-120.8,urcrnrlat=37.5)

# draw lat/lon grid lines
map.drawparallels(np.arange(36.4, 38, 0.5), labels=[1,0,0,0], linewidth=0)
map.drawmeridians(np.arange(-123, -120, 0.5), labels=[0,0,0,1], linewidth=0)
        
###################################plot faults###################################
faults = read_faults('calif-fault.gmtlin')
for i in faults:
    faults_lons = faults[i][0]
    faults_lats = faults[i][1]
    x_,y_ = map(faults_lons, faults_lats)
    map.plot(x_,y_,'b')
    
###################################plot aftershocks###################################
x, y = map(lons, lats)

#scale the size
min_size = 30
max_size = 60
min_mag = min(mags)
max_mag = max(mags)

frac = [(_i - min_mag) / (max_mag - min_mag) for _i in mags]
magnitude_size = [(_i * (max_size - min_size)) **2 for _i in frac]
map.scatter(x, y, marker='o', s=magnitude_size, c='r',
            zorder=10)

###################################plot main event###################################
frac = [(main_mag - min_mag) / (max_mag - min_mag) for _i in mags]
main_size = [(_i * (max_size - min_size)) **2 for _i in frac]
#print main_lon, main_lat
_x, _y = map(main_lon, main_lat)
map.scatter(_x, _y, marker='*', s=main_size, c='green',
                    zorder=10)

###################################plot scale, annotation, and legend###################################
map.drawmapscale(-122.4, 36.32, -122, 37, 30, barstyle='fancy')
plt.annotate('Berkeley Seismological Lab', xy=(0.01, 0.01), xycoords='axes fraction', fontsize = 16)
legends = []
for mag_ in [2.,3.,4.,5.,6.,7.]:
    frac = (mag_ - min_mag) / (max_mag - min_mag)
    magnitude_size = (frac * (max_size - min_size)) **2
    l1 = plt.scatter([], [], marker='o', s=magnitude_size, c='red', zorder=10)
    legends.append(l1)
    
ll = ['M2\n', 'M3\n', 'M4\n', 'M5\n', 'M6\n', 'M7\n']

leg = plt.legend(legends, ll, scatterpoints = 1, ncol=6, frameon=False, fontsize=12, handlelength=7, \
    columnspacing = 0.01, bbox_to_anchor=(0.99, -0.06))    
    
x, y = map(main_lon - 0.042, main_lat - 0.042)
x1, y1 = map(main_lon -0.28, main_lat -0.15)
plt.annotate('10/18/1989\n     M 6.9', xy=(x, y), xytext=(x1, y1), arrowprops=dict(facecolor='black', shrink=0.05, width = 0.5,
frac = 0.2, headwidth = 7),
zorder=10) 

###################################plot TOPO background###################################
ls = LightSource(azdeg = 90, altdeg = 60)
#read data from GMT grid file
etopodata = Dataset('./srtm_12_05.tif.grd')

#get lat, lon, and topo
topoin = etopodata.variables['z'][:]
lons_topo = etopodata.variables['x'][:]
lats_topo  = etopodata.variables['y'][:]

#set the negative elevation to 0
topoin[np.where(topoin<0)] = 0

# transform to nx x ny regularly spaced 5km native projection grid
nx = int((map.xmax-map.xmin)/90.)+1; ny = int((map.ymax-map.ymin)/90.)+1
topodat = map.transform_scalar(topoin,lons_topo ,lats_topo ,nx,ny)
# plot image over map with imshow.
rgb = ls.shade(topodat, cm.GMT_drywet)
im = map.imshow(rgb)

plt.show()
```
